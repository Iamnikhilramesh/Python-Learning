```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
t = list(range(0,20,1))
x = list(range(500,600,5))
y = list(range(20,40,1))
z = list(range(50,90,2))
```

```python
print(t)
print(x)
print(y)
print(z)
```

# Matplotlib 

```python
#scatter 
plt.scatter(x,y)
plt.title("Niky Scatter")
plt.legend("D")
plt.xlabel("X-label")
plt.ylabel("Y Label")
plt.show()
```

```python
#bar charts
plt.bar(x,y)
plt.title("Niky Bar Plot")
plt.legend("D")
plt.xlabel("X-label")
plt.ylabel("Y Label")
plt.show()
```

```python
#use of  subplot
plt.subplot(131)
plt.scatter(x,y)
plt.subplot(132)
plt.bar(x,y)
plt.subplot(133)
plt.plot(x,y)
```

```python
#annotation and grid
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict())
# GRid
plt.grid(True)
plt.ylim(-2, 2)
plt.show()
```
